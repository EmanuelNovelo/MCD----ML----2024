```python
import pandas as pd
```

```python
# Carga de datos
df = pd.read_csv('C:/Users/emanuel.novelo/Desktop/MCD - 2024-2026/ML - 2do Tetra/MCD----ML----2024/data/CO2_emission.csv')
df.head()
```

```python
# Tipo de datos por columna
df.info()
```

```python
# Asignando columna Model_Year como indice en un df copia
df_copy = df.copy()
df_copy = df_copy.set_index(['Model_Year'])
df_copy.head()
```

```python
# Creando columnas nuevas, modificando y eliminando columnas
# Añadiendo "L" de Litros a la columna "Engine_Size"
df['Engine_Size'] = df['Engine_Size'].astype(str) + 'L'

# Creando una nueva columna "Engine_characteristhics" a partir de "Engine_Size", "Cylinders", "Transmission"
df['Engine_characteristics'] = df.apply(lambda row: (row['Engine_Size'], row['Cylinders'], row['Transmission']), axis=1)

# Borrando columnas que ya fueron concatenadas
df = df.drop(['Engine_Size','Cylinders','Transmission'], axis=1)

df.head()
```

```python
# Filtrando registros y creando a su vez subconjuntos de datos y grupos
# Selecciondo vehiculos modelo 2021
df_2021 = df[df['Model_Year'] == 2021]
df_2021.head()

# Seleccionando vehiculos cuyo fabricante empieza con la letra B
df_B_make = df[df['Make'].str.startswith('B')]
df_B_make.head()
```

```python
# Creando subconjuntos y grupos de datos
# Creando un diccionario donde la "llave" es por Marca o "Make", es decir los datos se guardan por subconjuntos...
# dados por la marca que fabrico los vehículos

cars_dict = {make: df_group for make, df_group in df.groupby('Make')}
cars_dict
```
